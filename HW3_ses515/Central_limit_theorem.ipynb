```python
import numpy as np
import matplotlib.pyplot as pl
%pylab inline 
```

___

# Plot distributions

### Normal

```python
mean = 12
stdv = 10

np.random.seed(14)

for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.normal(mean, stdv, ran_size)
    pl.plot(ran_size,sample.mean(), "o",color="red")
    
pl.xlabel('Sample Size', fontsize=14)    
pl.ylabel('Sample Means', fontsize=14)
pl.title('Normal Distribution Means', fontsize=18)    
```

___

# Poisson

```python
lamda= 12

for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.poisson(lamda, ran_size)
    pl.plot(ran_size, sample.mean(), "o", color="blue");
    
pl.xlabel('Sample Size', fontsize=14)    
pl.ylabel('Sample Means', fontsize=14)
pl.title('Poission Distribution Means', fontsize=18)      
```

```python
k = 12
n =24
p =.5

for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.binomial(n, p, ran_size)
    pl.plot(ran_size, sample.mean(), "o",color="k");
    
pl.xlabel('Sample Size', fontsize=14)    
pl.ylabel('Sample Means', fontsize=14)
pl.title('Binomial Distribution Means', fontsize=18)  
```

### Chi-squared

```python
k = 12

for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.chisquare(k, ran_size)
    pl.plot(ran_size, sample.mean(), "o", color="green");   
    

pl.xlabel('Sample Size', fontsize=14)    
pl.ylabel('Sample Means', fontsize=14)
pl.title('Chi-Squared Distribution Means', fontsize=18) 
```

___

### Exponential

```python
scale = 12

for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.exponential(scale, ran_size)
    pl.plot(ran_size, sample.mean(), "o",color="yellow");
    
    
pl.xlabel('Random Sample Size', fontsize=14)    
pl.ylabel('Sample Means', fontsize=14)
pl.title('Exponential Distribution Means', fontsize=18)
```

```python
# set variables needed in all equations. We chose these numbers so that 
# each of the distributions would have the same mean. 

mean = 12
stdv = 10
lamda= 12
scale = 12
k = 12
n =24
p =.5

# define seed for all distributions
np.random.seed(14)

# Normal Distribution
for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.normal(mean, stdv, ran_size)
    pl.plot(ran_size,sample.mean(), "o",color="red");
    
# Poission Distribution
for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.poisson(lamda, ran_size)
    pl.plot(ran_size, sample.mean(), "o", color="blue");
    
# Chi square Distribution 
for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.chisquare(k, ran_size)
    pl.plot(ran_size, sample.mean(), "o", color="green");   
    
#Binomial distribution
for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.binomial(n, p, ran_size)
    pl.plot(ran_size, sample.mean(), "o",color="k");
    
# Exponential distribution
for i in range(100):
    ran_size = np.random.randint(10,2000)
    sample = np.random.exponential(scale, ran_size)
    pl.plot(ran_size, sample.mean(), "o",color="yellow");
    
    
pl.xlabel('Sample Size', fontsize=14)    
pl.ylabel('Sample Means', fontsize=14)
pl.title('Distribution of Means', fontsize=18)
```

# Interpretation
These plots show that as the sample size gets larger, the variation in the mean gets smaller, getting closer to the population's true mean.


# Histogram

```python
# calculating means, adding them to list
means = []

# define seed for all distributions
np.random.seed(14)

# Normal Distribution
for i in range(100):
    norm_ran = np.random.randint(10,2000)
    sample = np.random.normal(mean, stdv, norm_ran)
    means.append(sample.mean())
    
# Poission Distribution
for i in range(100):
    poi_rand = np.random.randint(10,2000)
    sample = np.random.poisson(lamda, poi_rand)
    means.append(sample.mean())
    
# Chi square Distribution 
for i in range(100):
    chi_rand = np.random.randint(10,2000)
    sample = np.random.chisquare(k, chi_rand)
    means.append(sample.mean())  
    
#Binomial distribution
for i in range(100):
    bio_rand = np.random.randint(10,2000)
    sample = np.random.binomial(n, p, bio_rand)
    means.append(sample.mean())
     
#Expo distribution        
for i in range(100):
    expo_rand = np.random.randint(10,2000)
    sample = np.random.exponential(scale, expo_rand)
    means.append(sample.mean())        
        
pl.hist(means, bins=100);    
pl.xlabel('Sample Means', fontsize=14)    
pl.ylabel('Mean count', fontsize=14)
pl.title('Distribution of Means', fontsize=18)
pl.xlim(10, 14)
```

```python
# normalize and add fit line. 
# source: https://stackoverflow.com/questions/20011494/plot-normal-distribution-with-matplotlib

import scipy.stats as stats

sortedmeans = sorted(means)  
fit = stats.norm.pdf(sortedmeans, np.mean(sortedmeans), np.std(sortedmeans))
pl.plot(sortedmeans,fit)
pl.hist(means, bins=90, normed=True) 
pl.title('Distribution of Means Normalized', fontsize=18)
pl.xlabel('Sample Means', fontsize=14)    
pl.ylabel('Mean count', fontsize=14)
pl.show() 
```

# Interpretation
Similar to the scatter plot, this histogram shows that the means are normally distributed, just as we'd expect from the central limit theorem.
